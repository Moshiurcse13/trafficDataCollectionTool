```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob
import os

lower_red = np.array([0,150,150])
upper_red = np.array([10,255,255])

lower_orange = np.array([10,100,100]) 
upper_orange = np.array([15,255,255])

lower_green=np.array([60, 110,100])
upper_green=np.array([70, 255, 255])

images = []
 

f= open("june(21_30).csv","a+")
Colored_Pixels = []
imgNo=0

for img in glob.glob("C:/Users/Tazin/Documents/Thesis_TFP/june after 20/crop/mask/*.png"):            
    print("Reading image no ",imgNo)
    imgNo+=1

    inputImage = cv2.imread(str(img))
    image_name=os.path.basename(img)
    arr = str(image_name).split('.')
    arr1 = str(arr[0]).split('_')
   
    f.write('\n') 
    f.write(arr1[3])
    f.write(",")
    f.write(arr1[4])
    f.write(",")
    f.write(arr1[0])
    f.write(",")
    f.write(arr1[1])
    f.write(",")
    f.write(arr1[2])
    
    f.write(",")
    if(arr1[0]=='Fri' or arr1[0]=='Sat'):
        f.write("1,")
    else:
        f.write("0,")
        
    hsv = cv2.cvtColor(inputImage, cv2.COLOR_BGR2HSV)
    mask_red =  cv2.inRange(hsv, lower_red, upper_red)
    mask_orange = cv2.inRange(hsv, lower_orange, upper_orange)
    mask_green = cv2.inRange(hsv, lower_green, upper_green)
    inputImageGray = cv2.cvtColor(inputImage, cv2.COLOR_BGR2GRAY)
    seg = 1
    height, width, channels = inputImage.shape
    red = 0
    orange = 0
    green = 0
    black=0
    for i in range (0,height,30):
          for j in range (0, width,50):
                for k in range(i,i+23,1):
                    for l in range(j,j+25,1):
                        if(mask_red[k,l]>0):
                            red = red+1
                            if(imgNo==1):
                                Colored_Pixels.append([k,l])
                        if(mask_orange[k,l]>0):
                            orange = orange+1
                            if(imgNo==1):
                                Colored_Pixels.append([k,l])
                        if(mask_green[k,l]>0):
                            green = green+1
                            if(imgNo==1):
                                Colored_Pixels.append([k,l])
                        if(inputImageGray[k][l]==0):
                            black = black+1
                            if(imgNo==1):
                                Colored_Pixels.append([k,l])
                        
                if(red>green and red>orange and 575>black):
                    f.write("3,")
                elif(orange>green and orange>red and 575>black):
                    f.write("2,")
                elif(red<green and green>orange and 575>black):
                    f.write("1,") 
                elif(black==575):
                    f.write("0,")
                red = 0
                orange = 0
                green = 0
                black=0
                
    seg = seg+1
            
f.close()
```
